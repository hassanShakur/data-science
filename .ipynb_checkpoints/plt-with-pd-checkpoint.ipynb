```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Stock graph
ts = pd.Series(
    data=np.random.randn(100),
    index = pd.date_range('1/1/2023', periods=100)
)

# Calc cumulative sums
ts = ts.cumsum()
ts.plot()
```

```python
heart_data = pd.read_csv('data/heart_desease.csv')
over_50 = heart_data[heart_data.age > 50]
over_50.head()
```

```python
fig, ax = plt.subplots()

scatter = ax.scatter(
    x=over_50.age,
    y=over_50.chol,
    c=over_50.target # for color in this case
)

ax.set(
    label='Heart Data & Cholesteral',
    xlabel='Age',
    ylabel='Cholesteral Level'
)

# Add a legend
# legend_elements are extracted from the 'c' param when creating the scatter
ax.legend(*scatter.legend_elements(), title='Target')

# Add horizontal line for mean
ax.axhline(over_50.chol.mean(), linestyle='-.')
```

## Colors

```python
fig, ax = plt.subplots()

scatter = ax.scatter(
    x=over_50.age,
    y=over_50.chol,
    c=over_50.target,
    cmap='winter' # Change color scheme
)

ax.set(
    label='Heart Data & Cholesteral',
    xlabel='Age',
    ylabel='Cholesteral Level'
)

# Add a legend
# legend_elements are extracted from the 'c' param when creating the scatter
ax.legend(*scatter.legend_elements(), title='Target')

# Add horizontal line for mean
ax.axhline(over_50.chol.mean(), linestyle='-.')
```
