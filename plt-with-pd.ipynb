```python
heart_data = pd.read_csv('data/heart_desease.csv')
over_50 = heart_data[heart_data.age > 50]
over_50.head()
```

```python
fig, ax = plt.subplots()

# c is for color in this case
scatter = ax.scatter(
    x=over_50.age,
    y=over_50.chol,
    c=over_50.target
)

ax.set(
    label='Heart Data & Cholesteral',
    xlabel='Age',
    ylabel='Cholesteral Level'
)

# Add a legend
# legend_elements are extracted from the 'c' param when creating the scatter
ax.legend(*scatter.legend_elements(), title='Target')

# Add horizontal line for mean
ax.axhline(over_50.chol.mean(), linestyle='-.')
```
