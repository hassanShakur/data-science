## Recommended way of creating figures

The recommended is using the API OOP method when creating `figures`. Here the figure is the general structure while the `subplots` are axes in the figure and lines.

### Line Graphs

```python
# Initialize the instance
fig, ax = plt.subplots(figsize=(5, 5))

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Actual plot
ax.plot(x, y)

# Set axes and title
ax.set(title='A Linear Function',
       xlabel='x-axis',
       ylabel='y-axis'
      )

# Exporting
# fig.savefig('path/to/save')
```

```python
# Create linearly spaced nums btwn 2 vals
arr = np.linspace(0, 10, 50)
arr[:10]
```

```python
# Plot
fig, ax = plt.subplots()
ax.plot(arr, arr**2)
```

### Scatter Plots

```python
# Scatter
fig, ax = plt.subplots()
ax.scatter(arr, np.exp(arr))
```

```python
# Sin wave
fig, ax = plt.subplots()
ax.scatter(arr, np.sin(arr))
```

```python
books = {
    'Educational': 22,
    'Romance': 43,
    'Explorative': 10,
    'Poetic': 30,
    'Perception': 7
}
```

### Bar Graphs

```python
# Bar
fig, ax = plt.subplots()
ax.bar(books.keys(), books.values())

ax.set(
    title = 'My Books',
    ylabel = 'Count'
)
```

```python
# Horizontal bar - Use Lists
fig, ax = plt.subplots()
ax.barh(list(books.keys()), list(books.values()))
```

```python
# Create a normal Gauss probability distribution data
pdist = np.random.randn(1000)
pdist[:10]
```

```python
# Histogram
fig, ax = plt.subplots()
ax.hist(pdist)
```

### Subplots

#### Version 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 10)
)

ax1.plot(arr)
ax2.scatter(np.random.random(40), np.random.random(40))
ax3.hist(pdist)
ax4.bar(books.keys(), books.values())

# Add General figure title
fig.suptitle('Working with Subplots', fontsize=18, fontweight='bold');
```
